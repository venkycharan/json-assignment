# JSON Assignment
This notebook demonstrates building, parsing, and manipulating JSON data using Python.

```python
import json  # built-in module for working with JSON
```

```python
## Task 1 — Build a JSON Structure
user_profile = {
    "name": "Venky",
    "age": 25,
    "email": "venky@example.com",
    "is_active": True,
    "skills": ["Python", "Machine Learning", "Networking"]
}

print(json.dumps(user_profile, indent=4))
```

```python
## Task 2 — Parse an API Response
api_response = '{"status": "success", "data": {"user_id": 101, "username": "alex99", "score": 87.5}}'
parsed = json.loads(api_response)
username = parsed["data"]["username"]
score = parsed["data"]["score"]
print("Username:", username)
print("Score:", score)
print(f"User {username} scored {score} points")
```

```python
## Task 3 — Handle Nested JSON
nested = {
    "name": "Priya",
    "address": {
        "city": "Bengaluru",
        "state": "Karnataka",
        "zip": "560001"
    }
}

print("City:", nested["address"]["city"])
print("Zip Code:", nested["address"]["zip"])

nested["address"]["country"] = "India"
print(json.dumps(nested, indent=4))
```
